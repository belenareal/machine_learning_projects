#### EXPLORATORY DATA ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('data/predict_future_sales/sales_train.csv')
df
```

```python
print(f'Number of null values in the dataset: {df.isnull().values.any().sum()}')
print(f'Number of duplicated rows: {df.duplicated().sum()}')

df = df.drop_duplicates()
```

```python
df_items = pd.read_csv('data/predict_future_sales/items.csv')
df_items.head()
```

```python
df_item_cat = pd.read_csv('data/predict_future_sales/item_categories.csv')
df_item_cat.head()
```

```python
df_shops = pd.read_csv('data/predict_future_sales/shops.csv')
df_shops.head()
```

```python
df = pd.merge(df, df_items, on='item_id', how='left')
df = pd.merge(df, df_item_cat, on='item_category_id', how='left')
df = pd.merge(df, df_shops, on='shop_id', how='left')
```

```python
df['date'] = pd.to_datetime(df['date'])
df_sorted = df.sort_values(by='date')
```
