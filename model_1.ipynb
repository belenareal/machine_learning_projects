```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_log_error
```

```python
df = pd.read_csv('data/data_ids.csv')
```

```python
df
```

```python
train = df[(df['year'] == 0) | ((df['year'] == 1) & (df['month'] <= 7))]
test = df[(df['year'] == 1) & (df['month'] > 7)]
X_columns = ['year', 
             'month', 
             'hour', 
             'holiday', 
             'weekday', 
             'workingday', 
             'temp',  
             'humidity', 
             'windspeed', 
             'clear', 
             'heavy_rain', 
             'misty', 
             'rain', 
             'fall', 
             'spring', 
             'summer', 
             'winter',
             'peak_hour', 
             'peak_month']
y_columns = ['y']

X_train = train[X_columns]
y_train = train[y_columns]
X_test = test[X_columns]
y_test = test[y_columns]
```

```python
model_lgb = LGBMRegressor()
model_lgb.fit(X_train, np.log(y_train))
y_pred_lgb = model_lgb.predict(X_test)

y_pred_lgb = np.exp(y_pred_lgb)
lgb_msle = mean_squared_log_error(y_test, y_pred_lgb)

print(f'MSLE (LightGBM): {lgb_msle}')
```
