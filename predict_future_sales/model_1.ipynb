```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_log_error, mean_squared_error
```

```python
df = pd.read_csv('data/data_group_price.csv')
```

```python
df
```

```python
df.columns
```

```python
train = df[df['date_block_num'] < 32]
test = df[df['date_block_num'] >= 32]
X_columns = ['date_block_num', 
             'shop_id', 
             'item_id', 
             'item_price_mean', 
             'item_price_median', 
             'item_price_std',
             'item_price_min', 
             'item_price_max', 
             'min_max']
y_columns = ['item_cnt_day_sum']

X_train = train[X_columns]
y_train = train[y_columns]
X_test = test[X_columns]
y_test = test[y_columns]
```

```python
model_lgb = LGBMRegressor()
model_lgb.fit(X_train, y_train)
y_pred = model_lgb.predict(X_test)

y_pred = np.exp(y_pred)
lgb_mse = mean_squared_error(y_test, y_pred)

print(f'MSE (LightGBM): {lgb_mse}')
```
